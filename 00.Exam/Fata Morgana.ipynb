# Fata Morgana

### 1. Introduction
Fata Morgana is an atmospheric optical phenomenon that causes complex and distorted images of distant objects. Result of light refraction through layers of air with different temperatures. When warm air lies above colder air, light bends and creates images that appear at different locations from their actual positions. The name originates from Italian folklore about the sorceress Morgana le Fay, sister of King Arthur. The phenomenon occurs due to temperature gradients in the atmosphere, which alter the path of light, creating illusions of distant objects.

### 2. Physical Foundations
    
    --The refraction of light is described by Snell's Law:
<div style="text-align: center; font-size: 19px;">
    \( n_1 \sin(\theta_1) = n_2 \sin(\theta_2) \)
</div>
    
    --where:
<div style="text-align: center; font-size: 19px;">
    \( n_1\) and \( n_2\) are the refractive indices of the two media.
</div>
<div style="text-align: center; font-size: 19px;">
    \(\theta_1\) and \(\theta_2\) are the angles of incidence and refraction.
</div>
    
    Snell's Law can be used to explain how light behaves when it transitions between two media with different refractive indices. This is crucial for understanding Fata Morgana, as the atmosphere consists of layers with varying temperatures and densities.
    
    Snell's Law is used to calculate the angle of refraction when light passes from one medium to another:

<div style="text-align: center; font-size: 19px;">
   \[ \theta_2 = \arcsin \left( \frac{n_2}{n_1} \sin(\theta_1) \right) \]

</div>

### 3. Refraction of Light in the Atmosphere
    Temperature Gradients and Refractive Index
    Temperature gradients in the atmosphere lead to changes in the refractive index, which in turn causes light refraction. This relationship can be described by the following formula:

<div style="text-align: center; font-size: 19px;">
    \[ n(z) = 1 + \beta (T(z) - T_0) \]
</div

        --where
<div style="text-align: center; font-size: 19px;">
    \(n(z)\) is the refractive index at altitude \(z\)
</div>-
<div style="text-align: center; font-size: 19px;">
    \(\beta\)  is the temperature coefficient
</div>-
<div style="text-align: center; font-size: 19px;">
    \(T(z)\) is the temperature at altitude \(z\)
</div>-
<div style="text-align: center; font-size: 19px;">
    \(T_0\) is the reference temperature
</div>

    Equations for Refraction in the Atmosphere
    Changes in the refractive index with altitude can be described by the differential equation:
    
<div style="text-align: center; font-size: 19px;">
    \[ \frac{d\theta}{dz} = \frac{1}{n} \frac{dn}{dz} \]
</div>

    --where
<div style="text-align: center; font-size: 19px;">
    \(\theta\) is the angle of the light ray
</div>-
<div style="text-align: center; font-size: 19px;">
    \(z\) is the altitude
</div>-
<div style="text-align: center; font-size: 19px;">
    \(n\) is the refractive index
</div>

### 4. At Sea 
    Historical records of the Fata Morgana date back to ancient times. Descriptions of this phenomenon can be found in written sources from ancient China, Egypt, and Greece. Mariners often mentioned the Fata Morgana in their logs, describing ghostly ships and deceptive coastlines. Sea Fata Morganas are among the most commonly encountered. They can create illusions of flying ships or ghostly shores, confusing sailors and giving rise to legends of mysterious phenomena at sea. Over the years, numerous well-documented observations of Fata Morgana have been recorded. One of the most famous examples is the case of the "Flying Dutchman" - a legendary ghost ship that has often been associated with Fata Morgana.

### 5. In the Mountains
    When this optical phenomenon occurs in the mountains, it creates surreal and distorted images of distant peaks, valleys, and landscapes, often leading to misconceptions and myths. In mountainous areas, Fata Morgana forms due to the refraction of light through layers of air with different temperatures. Temperature gradients in these regions can be quite significant due to varying altitudes and terrain features. These temperature differences cause changes in the air's refractive index, which bends the light and creates deceptive images. In mountainous conditions, layers of warm and cold air can alternate rapidly, resulting in more complex and dynamic forms of Fata Morgana. These conditions often occur in the early morning or late evening hours when temperature changes are most pronounced. The distorted images created by Fata Morgana in the mountains can confuse observers, such as peaks appearing higher or lower than they actually are, or valleys that look like lakes. This phenomenon can be so realistic that even experienced mountaineers can be deceived. Although Fata Morgana in the mountains is less common compared to its marine and desert versions, it adds another layer of mystery and beauty to already impressive mountain landscapes. These optical illusions highlight the complexity of natural phenomena and the need for careful observation and understanding of the environment.

```python
def refractive_index(temp):
    return 1 + 0.00029 * (temp - 20)

def temp_profile(z):
    return 20 - 0.01 * z + 0.00001 * z**2

z = np.linspace(0, 1000, 500)

temp_gradient = temp_profile(z)
n_gradient = refractive_index(temp_gradient)

theta0 = np.radians(0.5)

theta = np.zeros_like(z)
theta[0] = theta0

for i in range(1, len(z)):
    dz = z[i] - z[i - 1]
    dndz = (n_gradient[i] - n_gradient[i - 1]) / dz
    theta[i] = theta[i - 1] + (dndz / n_gradient[i]) * dz

plt.figure(figsize=(10, 6))
plt.plot(z, np.degrees(theta), color='purple', linewidth=2)
plt.fill_between(z, np.degrees(theta), color='lavender')
plt.xlabel('Height (m)', fontsize=14, fontweight='bold')
plt.ylabel('Angle of Refraction (Â°)', fontsize=14, fontweight='bold')
plt.title('Light Refraction at Different Heights', fontsize=16, fontweight='bold')
plt.grid(True)
plt.gca().set_facecolor('lightgray')
plt.show()
```

### 6. In deserts 
    Fata Morganas often appear in the form of oases that appear real but are actually just reflections of the sky on heated sand. Appears under specific meteorological conditions when the temperature differences between different layers of air are greatest. Sometimes these conditions can be predictable, allowing observers to anticipate the phenomenon. These illusions can be very realistic and deceive travelers."

```python
def refractive_index(temp):
    return 1 + 0.00029 * (temp - 20)

def temp_profile(z):
    return 20 - 0.01 * z + 0.00001 * z**2

z = np.linspace(0, 1000, 500)

temp_gradient = temp_profile(z)
n_gradient = refractive_index(temp_gradient)

theta0 = np.radians(0.5) 

theta = np.zeros_like(z)
theta[0] = theta0

for i in range(1, len(z)):
    dz = z[i] - z[i - 1]
    dndz = (n_gradient[i] - n_gradient[i - 1]) / dz
    theta[i] = theta[i - 1] + (dndz / n_gradient[i]) * dz

plt.figure(figsize=(10, 6))
plt.plot(z, np.degrees(theta), color='green', linewidth=2, linestyle='-', marker='o', markersize=5, markerfacecolor='yellow', label='Angle of Refraction')

plt.xlabel('Height (m)', fontsize=14, fontweight='bold', color='navy')
plt.ylabel('Angle of Refraction (Â°)', fontsize=14, fontweight='bold', color='navy')
plt.title('Light Refraction with Complex Temperature Profile', fontsize=16, fontweight='bold', color='navy')
plt.grid(True, linestyle='--', linewidth=0.8, color='gray')
plt.legend(fontsize=12, loc='upper right')

plt.gca().set_facecolor('#f0f0f0')

plt.tight_layout()
plt.show()
```

### 8. In Astronomy
    The Fata Morgana phenomenon occurs when the atmosphere of a planet or moon creates optical illusions similar to those observed on Earth. These illusions arise due to the refraction of light through layers of air with different temperatures and densities. In astronomy, these effects can be observed not only on our planet but also in outer space, especially on planets with atmospheres such as Mars and Venus.

    Mars has a relatively thin atmosphere composed mainly of carbon dioxide. However, temperature differences in Mars' atmosphere can create optical illusions. For example, observations from Earth-based telescopes or Martian orbiters can be affected by these effects, creating the illusion of objects or structures that do not exist on the planet's surface.

    Venus has a very dense and hot atmosphere, primarily composed of carbon dioxide and clouds of sulfuric acid. These conditions can lead to strong light refraction and the creation of optical illusions resembling the Fata Morgana. Observations of Venus' surface are often hindered by these atmospheric phenomena.

    The Fata Morgana can significantly impact astronomical observations. Optical illusions like these can distort images obtained from telescopes or space probes, leading to incorrect interpretations of data and erroneous conclusions about the characteristics of observed objects. Observations of the Fata Morgana and other optical phenomena can provide valuable information about the structure and composition of other planets' atmospheres. They help scientists understand how light refracts and scatters under different atmospheric conditions.

    Modern technologies, such as telescopes with adaptive optics, have been developed to compensate for the effects of atmospheric distortions. These technologies use sensors and mirrors that adjust in real-time to correct distortions caused by the atmosphere, enhancing the quality of images and the accuracy of observations.

### 9. Modeling Fata Morgana

    --Dependency of Refractive Index on Temperature
    --Visualization of the Dependency of Refractive Index on Temperature

```python
temperatures = np.linspace(-30, 30, 100)
beta = 0.00029
T0 = 20

def calculate_refractive_index(temp):
    return 1 + beta * (temp - T0)

n_values = calculate_refractive_index(temperatures)

plt.figure(figsize=(10, 6), facecolor='lightgrey')

plt.plot(temperatures, n_values, color='royalblue', linewidth=2, linestyle='-', marker='o', markersize=8, markerfacecolor='red', label='Refractive Index Dependency')

# Plot styling
plt.xlabel('Temperature (Â°C)', fontsize=14, fontweight='bold', color='darkblue')
plt.ylabel('Refractive Index', fontsize=14, fontweight='bold', color='darkblue')
plt.title('Dependency of Refractive Index on Temperature', fontsize=16, fontweight='bold', color='navy')
plt.grid(True, linestyle='--', linewidth=0.8, color='gray')
plt.legend(fontsize=12, loc='upper left')

plt.tick_params(axis='both', which='major', labelsize=12)
plt.axhline(y=1, color='gray', linestyle='--', linewidth=0.8)
plt.axvline(x=0, color='gray', linestyle='--', linewidth=0.8)

plt.tight_layout()
plt.show()
```

### 10. Conclusion:
